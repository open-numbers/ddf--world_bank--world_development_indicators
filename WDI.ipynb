# Concepts

```python
import re

def to_concept_id(s):
    '''convert a string to lowercase alphanumeric + underscore id for concepts'''
    return re.sub(r'[/ -\.]+', '_', s).lower()
```

```python
country['country'] = country['Country Code'].apply(to_concept_id)
series['series'] = series['Series Code'].apply(to_concept_id)
```

```python
concepts_discrete = pd.DataFrame([], columns=['concept', 'name', 'type'])

concepts_discrete['name'] = np.concatenate([country.columns, series.columns])
```

```python
concepts_discrete['concept'] = concepts_discrete['name'].apply(to_concept_id)
```

```python
concepts_discrete['type'] = 'string'
```

```python
concepts_discrete = concepts_discrete.append(pd.DataFrame([['year', 'Year', 'time']], index=[53], columns=concepts_discrete.columns))
```

```python
ed = [5, 7, 12, 31, 52, 51, 8, 13, 15, 19, 20, 21, 22, 41]

concepts_discrete['type'].iloc[ed] = 'entity domain'
```

```python
concepts_discrete.to_csv('./output/ddf--concepts--discrete.csv', encoding='utf-8', index=False)
```

```python
concepts_discrete
```

```python
concepts_continuous = series
```

```python
concepts_continuous['type'] = 'measure'
```

```python
idxs = np.r_[concepts_continuous.columns[-2:], concepts_continuous.columns[:-2]]
```

```python
concepts_continuous = concepts_continuous.loc[:, idxs]
```

```python
concepts_continuous.columns = list(map(to_concept_id, concepts_continuous.columns))
```

```python
def simple_df(s1, ser):
    da = ser[s1].dropna().unique()
    df = pd.DataFrame([], columns=[ 'name'])
    df['name'] = da
    df = df.reset_index()
    df.columns = [to_concept_id(s1), 'name']
    
    return df
```

```python
simple_df('License Type', series)
```

```python
simple_df('Periodicity', series).to_csv('./output/ddf--entities--periodicity.csv', index=False)
```
